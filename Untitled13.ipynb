# SPLITTING DATA INTO TRAINING AND TESTING SET

```python
weather_y=weather_df.pop("tempC")
weather_x=weather_df
```

# Applying linear regression model (examples : mintempC HeatIndexC and pressure)

```python
x_train, x_test , y_train, y_test = train_test_split(weather_x['mintempC'] , weather_y , test_size= 0.2, random_state=2)
x_train=np.array(x_train).reshape(-1,1)
x_test=np.array(x_test).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
model = LinearRegression()
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
plt.scatter(x_test,y_test,color="r")
plt.title("Linear Regression")
plt.ylabel("tempC")
plt.xlabel("mintempC")
plt.plot(x_test,model.predict(x_test),color="k")
plt.show()
```

```python
print("MSE", mean_squared_error(y_test,y_pred))
print("R squared", metrics.r2_score(y_test,y_pred))
```

```python
x_train, x_test , y_train, y_test = train_test_split(weather_x['HeatIndexC'] , weather_y , test_size= 0.2, random_state=2)
x_train=np.array(x_train).reshape(-1,1)
x_test=np.array(x_test).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
model = LinearRegression()
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
plt.scatter(x_test,y_test,color="r")
plt.title("Linear Regression")
plt.ylabel("tempC")
plt.xlabel("HeatIndexC")
plt.plot(x_test,model.predict(x_test),color="k")
plt.show()
```

```python
print("MSE", mean_squared_error(y_test,y_pred))
print("R squared", metrics.r2_score(y_test,y_pred))
```

```python
x_train, x_test , y_train, y_test = train_test_split(weather_x['pressure'] , weather_y , test_size= 0.2, random_state=2)
x_train=np.array(x_train).reshape(-1,1)
x_test=np.array(x_test).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
model = LinearRegression()
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
plt.scatter(x_test,y_test,color="r")
plt.title("Linear Regression")
plt.ylabel("tempC")
plt.xlabel("pressure")
plt.plot(x_test,model.predict(x_test),color="k")
plt.show()
```

```python
print("MSE", mean_squared_error(y_test,y_pred))
print("R squared", metrics.r2_score(y_test,y_pred))
```

# Applying multiple-linear regression


```python
train_X,test_X,train_y,test_y=train_test_split(weather_x,weather_y,test_size=0.2,random_state=2)
model.fit(train_X , train_y)
y_pred1 = model.predict(test_X)
print("MSE", mean_squared_error(test_y,y_pred1))
print("R squared", metrics.r2_score(test_y,y_pred1))
```

We note that the R squared and MSE metric has improved from linear to multiple linear regression

# Decision Tree Regression


```python
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(train_X,train_y)
prediction=regressor.predict(test_X)
```

```python
print("MSE", mean_squared_error(test_y,prediction))
print("R squared", metrics.r2_score(test_y,prediction))
```

# Random Forest Regression

```python
regr=RandomForestRegressor(max_depth=90,random_state=0,n_estimators=100)
regr.fit(train_X,train_y)
prediction1=regr.predict(test_X)
```

```python
print("MSE", mean_squared_error(test_y,prediction1))
print("R squared", metrics.r2_score(test_y,prediction1))
```

The Random Forest Regression is the more accurate model (has a low MSE and a good R near to 1)
